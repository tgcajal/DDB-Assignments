### Movie Recommendation System

By using the movie.csv and rating.csv data, and python(hybrid) code, select a user who has watched min 150 movies. By using that user's ID, create a hybrid recommender and recommend 10 movies. 

### Loading the datasets

```python
import pandas as pd

# dataset containing movieId, movie name and genre information of the movie
movie = pd.read_csv('movie.csv')
movie.head()
```

```python
# Data set containing UserID, movie name, movie vote and time information
rating = pd.read_csv('rating.csv')
rating.head()
```

```python
rating["userId"].nunique()
```

```python
# Only the id's of the movies voted by the users in the rating are available.
# We add the movie names and genre of the ids from the movie dataset.

df = movie.merge(rating, how="left", on="movieId")
df.head()
```

```python
df.shape
```

```python
# We calculate the total number of votes for each movie
comment_counts = pd.DataFrame(df["title"].value_counts())
comment_counts
```

```python
comment_counts.shape
```

```python
# We keep the names of movies with less than 1000 votes in rare_movies.
# And we subtract from the dataset

rare_movies = comment_counts[comment_counts["title"]<= 1000].index
rare_movies
```

```python
common_movies = df[~df["title"].isin(rare_movies)]
```

```python
common_movies.shape
```

```python
common_movies["title"].value_counts()
```

```python
# where userID's, movie names in the columns and their ratings as values in the index
# our pivot table for the dataframe.

user_movie_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
```

```python
user_movie_df.head()
```

```python
user_movie_df.shape
```

```python
# let's customize all functions in the config

def create_user_movie_df():
    import pandas as pd
    movie = pd.read_csv('movie.csv')
    rating = pd.read_csv('rating.csv')
    df = movie.merge(rating, how="left", on="movieId")
    comment_counts = pd.DataFrame(df["title"].value_counts())
    rare_movies = comment_counts[comment_counts["title"] <= 750].index
    common_movies = df[~df["title"].isin(rare_movies)]
    user_movie_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
    return user_movie_df
```

```python
user_movie_df = create_user_movie_df()
```

```python
user_movie_df.head()
```

```python
user_movie_df.columns
```

```python
rating.head()
```

### Selecting the user who has watched movies

```python
# Create a new variable

random_user = rating.groupby('userId').agg({'movieId':'count'})
random_user = random_user.reset_index()
random_user = random_user[random_user['movieId']>150]
random_user.sort_values('movieId', ascending=False)
```

```python
random_user = 111637
```

```python
random_user_df = user_movie_df[user_movie_df.index == random_user]
random_user_df.head()
```

```python
# We eliminate the movies they didn't vote for and list the movies they voted for

movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
movies_watched
```

```python
len(movies_watched)
```

### Accessing Data and Ids of Other Users Watching Same Movies

```python
# We select the columns of the movies watched by the user we selected from user_movie_df and create a new df.

movies_watched_df = user_movie_df[movies_watched]
movies_watched_df.head()
```

```python
user_movie_df.head()
```

```python
user_movie_df.info()
```

```python
# We calculate how many of the movies each user watched the selected user watched
# And we create a new df.

user_movie_count = movies_watched_df.T.notnull().sum()
user_movie_count.head()
```

```python
user_movie_count = user_movie_count.reset_index()
user_movie_count.head()
```

```python
user_movie_count.columns = ["userId", "movie_count"]
user_movie_count.head()
```

```python
user_movie_count[['movie_count']].sort_values('movie_count',ascending=False).head()
```

```python
user_movie_count[user_movie_count["movie_count"]>2]
```

```python
user_movie_count.describe()
```

```python
# We see that 60 percent of the movies voted by the option users can be seen by the viewers.

perc = len(movies_watched) * 60 / 100
users_same_movies = user_movie_count[user_movie_count["movie_count"] > perc]["userId"]
len(users_same_movies)
```

```python
perc
```

```python
# Identification of users most similar to the user to suggest

final_df = movies_watched_df[movies_watched_df.index.isin(users_same_movies)]
final_df.head()
```

```python
final_df.T
```

```python
final_df.corr()
```

```python
final_df.T.corr().unstack()
```

```python
corr_df = final_df.T.corr().unstack().sort_values()
corr_df.head()
```

```python
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.head()
```

```python
corr_df.index.names = ['user_id_1', 'user_id_2']
corr_df = corr_df.reset_index()
corr_df.head()
```

```python
corr_df[corr_df["user_id_1"] == random_user]
```

```python
top_users = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.63)][["user_id_2", "corr"]].reset_index(drop=True)
top_users = top_users.sort_values(by='corr', ascending=False)
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
```

```python
len(top_users)
```

```python
rating = pd.read_csv('rating.csv')
top_users_ratings = top_users.merge(rating[["userId", "movieId", "rating"]], how='inner')
top_users_ratings = top_users_ratings[top_users_ratings["userId"] != random_user]
top_users_ratings["userId"].unique()
```

```python
top_users_ratings.head()
```

### Calculating Weighted Average Recommendation Score and Keeping Top 5 Movies

```python
top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
top_users_ratings['weighted_rating'].head()
```

```python
top_users_ratings.head()
```

```python
top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
```

```python
recommendation_df = top_users_ratings.groupby('movieId').agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
recommendation_df.head()
```

```python
recommendation_df[recommendation_df["weighted_rating"] > 3]
```

```python
len(recommendation_df[recommendation_df["weighted_rating"] > 3])
```

```python
movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3].sort_values("weighted_rating", ascending=False)[0:10]
movies_to_be_recommend.head()
```

```python
movie = pd.read_csv('movie.csv')
movies_to_be_recommend.merge(movie[["movieId", "title"]])["title"]
```
